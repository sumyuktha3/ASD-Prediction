1. Input the data

```python
# Import data
import numpy as np
import pandas
from IPython.display import display

Adolescent_ASD = pandas.read_csv('https://github.com/Deakenttt/data_ASD/raw/main/Autism-Adolescent-Data.csv', na_values=['?'])
Adult_ASD = pandas.read_csv('https://github.com/Deakenttt/data_ASD/raw/main/Autism-Adult-Data.csv' , na_values=['?'])
Child_ASD = pandas.read_csv('https://github.com/Deakenttt/data_ASD/raw/main/Autism-Child-Data.csv', na_values=['?'])

display(Adolescent_ASD.head(n=5)) # display Autism-Adolescent data frame
print('Adolescent_ASD have {} samples'.format(Adolescent_ASD.shape[0]))
print('\t \t')
display(Adult_ASD.head(n=5)) # display Autism-Adult data frame
print('Adult_ASD have {} samples'.format(Adult_ASD.shape[0]))
print('\t \t')
display(Child_ASD.head(n=5)) # display Autism-Child data frame
print('Child_ASD have {} samples'.format(Child_ASD.shape[0]))

```

```python
from traitlets.traitlets import List
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
Adolescent_ASD.dropna(inplace = True)
print(Adolescent_ASD.shape[0])
#Adolescent_ASD.describe()
Adolescent_ASD_raw = Adolescent_ASD['Class/ASD']
Adolescent_features_raw = Adolescent_ASD[['age', 'gender', 'ethnicity', 'jundice', 'austim', 'contry_of_res', 'result',
                      'relation','A1_Score','A2_Score','A3_Score','A4_Score','A5_Score','A6_Score','A7_Score','A8_Score',
                      'A9_Score','A10_Score']]

scaler = MinMaxScaler()
numerical = ['age', 'result']
Adolescent_features_minmax_transform = pandas.DataFrame(data = Adolescent_features_raw)
Adolescent_features_minmax_transform[numerical] = scaler.fit_transform(Adolescent_features_raw[numerical]) # prep Age and Result use MinMaxScaler

Adolescent_features_final = pandas.get_dummies(Adolescent_features_minmax_transform)    # prep country
Adolescent_classes = Adolescent_ASD_raw.apply(lambda x: 1 if x == 'YES' else 0)

# Print the number of features after one-hot encoding
encoded = list(Adolescent_features_final.columns)

display(encoded)

# Uncomment the following line to see the encoded feature names

ADL_X_train, ADL_X_test, ADL_y_train, ADL_y_test = train_test_split(Adolescent_features_final.values, Adolescent_classes.values, test_size = 0.2, random_state = 1)
#display(Child_classes)

display(Adolescent_features_final)
print(Adolescent_features_final.values)
print("Training set has {} samples.".format(ADL_X_train.shape[0])) #--test--
print("Testing set has {} samples.".format(ADL_X_test.shape[0])) #--test--
print("data set has {} samples.".format(Adolescent_features_final.shape[0])) #--test--
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
knn_adu = KNeighborsClassifier(n_neighbors= 5)
knn_adu.fit(ADU_X_train,ADU_y_train)
#y_pred = knn.predict(ADU_X_test)
print(knn_adu.score(ADU_X_test,ADU_y_test))


knn_adl = KNeighborsClassifier(n_neighbors= 5)
knn_adl.fit(ADL_X_train,ADL_y_train)
print(knn_adl.score(ADL_X_test,ADL_y_test))


knn_chd = KNeighborsClassifier(n_neighbors= 5)
knn_chd.fit(CHD_X_train,CHD_y_train)
print(knn_chd.score(CHD_X_test,CHD_y_test))

knn = KNeighborsClassifier()
#best k value
from sklearn.model_selection._validation import cross_val_score
scores = cross_val_score(knn,ADU_X_train,ADU_y_train,cv=10,scoring='accuracy')
print(scores)
k_range = range(1,31)
k_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn,ADU_X_train,ADU_y_train,cv=10,scoring='accuracy')
    k_scores.append(scores.mean())
print(k_scores)

index = 1
for k in k_scores:
    if k == max(k_scores):
        print("Best k: ", index)
    index += 1
import matplotlib.pyplot as plt

plt.plot(k_range,k_scores)
plt.title('KNN performance in range k=(1,31)')
plt.xlabel('K value')
plt.ylabel('Cross_validation accuracy')
plt.show()


# Calculate accuracy, sensitivity, specificity 
def calculate (X_train, y_train, X_test, y_test):
  model = KNeighborsClassifier(n_neighbors=7)
  model.fit(X_train, y_train)
  y_predict = model.predict(X_test)
  tn, fp, fn, tp = confusion_matrix(y_test, y_predict).ravel()
  specificity = tn / (tn+fp)
  sensitivity = tp/(tp + fn)
  print("=====")
  print("Specificity score: ", round(specificity,4))
  print("Sensitivity score: ", round(sensitivity,4))
  print("Accuracy score: ", round( accuracy_score(y_test, y_predict), 4))

print("\nAdult")
calculate(ADU_X_train, ADU_y_train, ADU_X_test, ADU_y_test)
print("\nChilren")
calculate(CHD_X_train, CHD_y_train, CHD_X_test, CHD_y_test)
print("\nAdolescent")
calculate(ADL_X_train, ADL_y_train, ADL_X_test, ADL_y_test)
```

Plot learning curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve  
from sklearn.model_selection import ShuffleSplit


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
# Plot learning curve for Adult


title = r"Adult Learning Curves (KNN, n=7)"
cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)
estimator = KNeighborsClassifier(n_neighbors=7)
plot_learning_curve(
    estimator, title, Adult_features_final, Adult_classes, ylim=(0.7, 1.01), cv=cv, n_jobs=4
)
plt.show()
```

```python
# Plot learning curve for Children

title = r"Children Learning Curves (KNN, n=7)"
cv = ShuffleSplit(n_splits=50, test_size=0.2, random_state=0)
estimator = KNeighborsClassifier(n_neighbors=7)
plot_learning_curve(
    estimator, title, Child_features_final, Child_classes, ylim=(0.7, 1.01), cv=cv, n_jobs=4
)
plt.show()
```
